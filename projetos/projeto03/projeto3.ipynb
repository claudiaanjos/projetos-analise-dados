# Cenrio

O [Instacart](https://www.instacart.com/), um aplicativo estadunidense para pedidos e entrega de produtos de supermercado, tem como objetivo facilitar o abastecimento da geladeira e despensa com os produtos preferidos e itens essenciais dos seus clientes. Depois de selecionar os produtos pelo aplicativo, compradores do Instacart (que so contratados independentes) revisam o pedido, fazem as compras na loja e realizam a entrega.

A equipe de cincia de dados do Instacart desempenha um papel fundamental para proporcionar essa experincia de compra. Eles utilizam dados transacionais para desenvolver modelos que preveem quais produtos um usurio comprar novamente, experimentar pela primeira vez ou adicionar ao carrinho durante uma sesso de compras. Em 2017, o Instacart disponibilizou esses dados de forma aberta - confira o post deles no blog: [3 Million Instacart Orders, Open Sourced](https://tech.instacart.com/3-million-instacart-orders-open-sourced-d40d29ead6f2).

Como analista de dados seu objetivo  identificar associaes entre produtos que os clientes frequentemente compram juntos, ou seja, aplicar a Anlise de Cesta de Compras (Market Basket Analysis).

## Anlise de Cestas de Compras

A Anlise de Cesta de Compras, ou Market Basket Analysis (MBA)  uma tcnica de anlise de dados que faz uso de regras de associao para identificar os hbitos de compra dos clientes, fornecendo uma viso da combinao de produtos dentro das cestas de compras dos clientes analisados. Conhecer o perfil de compra do pblico-alvo  muito importante para aumentar o pontencial de recomendao, ajudando a melhorar as vendas.

Apesar de a nomenclatura Anlise de Cesta de Compras evocar uma imagem de carrinhos de compras e supermercados, cabe mencionar que existem outras reas em que essa abordagem pode ser empregada como, por exemplo, anlise de comprar em carto de crdito, anlise de padres em chamadas telefnicas, anlise de compra de servio telefnico, dentre outras aplicaes.

## Implementao do Algoritmo Apriori

O [algoritmo Apriori](https://pypi.org/project/efficient-apriori/) revela estruturas ocultas em dados categricos. O exemplo clssico  um banco de dados contendo compras de um supermercado. Cada compra est associada a vrios itens. Gostaramos de descobrir regras de associao, como {po, ovos} -> {bacon}, a partir dos dados. Esse  o objetivo do aprendizado de regras de associao, e o algoritmo Apriori  provavelmente o algoritmo mais famoso para esse problema.

Mtricas:

image.png

Exemplo:

image.png

Support = Frao de transaes que contm A e B.

image.png

Confidence = a frequncia com que B aparece em transaes que contm apenas A.

image.png

Lift = Quanto aumentou nossa confiana de que B ser comprado, dado que A foi comprado.

```python
# preparando a tupla com as transaes
transactions_tup = [tuple(row) for row in transactions['Products'].tolist()]
```

### Executando o Algoritmo Apriori Com Support = 0.01 e Confidence = 0.2

```python
st = datetime.now()
itemsets_ap, rules_ap = apriori(transactions_tup[:500000], min_support = 0.01, min_confidence = 0.2)
print('Tempo de Execuo : ', datetime.now() - st)
```

```python
rules_ap
```

```python
# Listas de itens para considerar
item_A = [27966,47209,21137,47766,21903,49683,47626,28204,16797,21903,21137,27966]
item_B = [13176,13176,24852,24852,24852,24852,24852,24852,24852,13176,13176,21137]
temp = pd.DataFrame()
temp['itemA'] = item_A
temp['itemB'] = item_B
```

```python
# Listas para as mtricas
support_A = []
support_B = []
support_AB = []
confidence_AB = []
lift_AB = []
```

```python
# Loop
for i in range(len(temp)):

    # Calcula o support de A
    support_A.append(itemsets_ap[1][tuple([temp['itemA'][i],])] / 500000)

    # Calcula o support de B
    support_B.append(itemsets_ap[1][tuple([temp['itemB'][i],])] / 500000)

    # Calcula o support de A e B
    if tuple([temp['itemA'][i],temp['itemB'][i]]) in itemsets_ap[2].keys():
        support_AB.append(itemsets_ap[2][tuple([temp['itemA'][i],temp['itemB'][i]])] / 500000)
    else:
        support_AB.append(itemsets_ap[2][tuple([temp['itemB'][i],temp['itemA'][i]])] / 500000)

    # Calcula a confidence
    confidence_AB.append(support_AB[i] / support_A[i])

    # Calcula o lift
    lift_AB.append(support_AB[i] / (support_A[i] * support_B[i]))
```
